```python
# Data exploration

Quick look at the chest X-ray dataset: image counts per class/split, and sample images.
```

```python
import sys, os
sys.path.append(os.path.abspath(".."))

import matplotlib.pyplot as plt
from PIL import Image

DATA_ROOT = os.path.join("..", "data", "raw", "lung-cancer-xray", "chest_xray_lung")
```

```python
# Count images per class per split
for split in ["train", "val", "test"]:
    split_dir = os.path.join(DATA_ROOT, split)
    for cls in os.listdir(split_dir):
        cls_dir = os.path.join(split_dir, cls)
        count = len(os.listdir(cls_dir))
        print(f"{split}/{cls}: {count} images")
```

```python
# Show a few sample images from each class
fig, axes = plt.subplots(1, 4, figsize=(12, 4))
sample_dirs = [
    os.path.join(DATA_ROOT, "train", "Cancer"),
    os.path.join(DATA_ROOT, "train", "NORMAL"),
]

i = 0
for cls_dir in sample_dirs:
    for fname in os.listdir(cls_dir)[:2]:
        img = Image.open(os.path.join(cls_dir, fname))
        axes[i].imshow(img, cmap='gray')
        axes[i].set_title(os.path.basename(cls_dir))
        axes[i].axis('off')
        i += 1

plt.tight_layout()
plt.show()
```
