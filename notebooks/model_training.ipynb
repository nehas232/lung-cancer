```python
import os, sys

current = os.path.abspath(os.getcwd())
project_root = current
for _ in range(5):
    if os.path.isdir(os.path.join(project_root, "src")):
        break
    project_root = os.path.dirname(project_root)

sys.path.insert(0, project_root)
print("Project root detected as:", project_root)
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

from src.config import DEVICE, BATCH_SIZE, IMAGE_SIZE
from src.models.cnn_model import LungCancerCNN
from src.models.train_model import train

DATA_ROOT = os.path.join(project_root, "data", "raw", "lung-cancer-xray", "chest_xray_lung")
print("Data root:", DATA_ROOT)
print("Exists:", os.path.isdir(DATA_ROOT))
```

```python
TRANSFORM = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),
    transforms.Resize(IMAGE_SIZE),
    transforms.ToTensor(),
])

train_dataset = datasets.ImageFolder(os.path.join(DATA_ROOT, "train"), transform=TRANSFORM)
val_dataset = datasets.ImageFolder(os.path.join(DATA_ROOT, "val"), transform=TRANSFORM)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

print(f"Train: {len(train_dataset)} images | classes: {train_dataset.classes}")
print(f"Val: {len(val_dataset)} images | classes: {val_dataset.classes}")
```

```python
device = DEVICE if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
model = LungCancerCNN().to(device)
```

```python
trained_model = train(model, train_loader, val_loader, device=device)
```
